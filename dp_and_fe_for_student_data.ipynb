```python
# ‡∏•‡∏ö‡∏´‡∏£‡∏∑‡∏≠‡πÅ‡∏ó‡∏ô‡∏Ñ‡πà‡∏≤‡∏ä‡∏∑‡πà‡∏≠‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡πÑ‡∏õ
df_master['‡∏ä‡∏∑‡πà‡∏≠'] = df_master['‡∏ä‡∏∑‡πà‡∏≠'].fillna('‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏')

# ‡πÅ‡∏ó‡∏ô‡∏Ñ‡πà‡∏≤‡∏≠‡∏≤‡∏¢‡∏∏‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢
df_master['‡∏≠‡∏≤‡∏¢‡∏∏'] = df_master['‡∏≠‡∏≤‡∏¢‡∏∏'].fillna(df_master['‡∏≠‡∏≤‡∏¢‡∏∏'].mean())

# ‡∏•‡∏ö‡∏Ñ‡πà‡∏≤ GPA ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥ (‡πÄ‡∏ä‡πà‡∏ô -1) ‡πÅ‡∏•‡∏∞‡πÅ‡∏ó‡∏ô‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢
df_master.loc[df_master['GPA'] < 0, 'GPA'] = None
df_master['GPA'] = df_master['GPA'].fillna(df_master['GPA'].mean())

# ‡∏£‡∏ß‡∏°‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô‡πÉ‡∏´‡πâ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏±‡∏ô
df_master['‡πÄ‡∏û‡∏®'] = df_master['‡πÄ‡∏û‡∏®'].replace({
    '‡∏ä‡∏≤‡∏¢': '‡∏ä‡∏≤‡∏¢', '‡∏ä': '‡∏ä‡∏≤‡∏¢', 'Male': '‡∏ä‡∏≤‡∏¢',
    '‡∏´‡∏ç‡∏¥‡∏á': '‡∏´‡∏ç‡∏¥‡∏á', '‡∏ç': '‡∏´‡∏ç‡∏¥‡∏á', 'Female': '‡∏´‡∏ç‡∏¥‡∏á'
})

# ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤ Missing ‡∏Ç‡∏≠‡∏á‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á‡πÅ‡∏•‡∏∞‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ)
df_master['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'] = df_master['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].fillna(df_master['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].mean())
df_master['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'] = df_master['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'].fillna(df_master['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'].mean())
```

```python
# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì BMI
df_master['BMI'] = df_master['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'] / ((df_master['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'] / 100) ** 2)

# ‡πÅ‡∏õ‡∏•‡∏á‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô datetime ‡πÅ‡∏•‡∏∞‡πÅ‡∏¢‡∏Å‡πÄ‡∏î‡∏∑‡∏≠‡∏ô/‡∏õ‡∏µ
df_master['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'] = pd.to_datetime(df_master['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'])
df_master['‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'] = df_master['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'].dt.month
df_master['‡∏õ‡∏µ‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'] = df_master['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏•‡∏á‡∏ó‡∏∞‡πÄ‡∏ö‡∏µ‡∏¢‡∏ô'].dt.year

# ---------------- ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÅ‡∏õ‡∏•‡∏á ----------------
print("\n‚úÖ ‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏±‡∏ö‡∏õ‡∏£‡∏∏‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•")
display(df_master.head(10))
```
